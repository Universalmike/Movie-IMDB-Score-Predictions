```python
import pandas as pd
import pickle
```

```python
June_Movies = pd.read_excel("Upcoming Movies and Shows in June (2024).xlsx")
June_Movies
```

```python

with open("movie-imdb-score-predictor", "rb") as f:
        model = pickle.load(f)
```

```python
June_Movies["imdb_scores_predictions"] = model.predict(June_Movies.drop(columns=["title", "released_date"]))
June_Movies
```

JUNE. 2024 MOVIES AND SHOWS IMDB_SCORE PREDICTIONS

```python
June_Movies["imdb_scores_predictions"] = June_Movies["imdb_scores_predictions"].round(1)
June_Movies
```

```python
June_Movies.to_excel('Top movies in June(2024) Movies imdb_score predictions.xlsx', index=False)
```
